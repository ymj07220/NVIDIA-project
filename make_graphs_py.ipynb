```python
import torch
print(torch.__version__)  # 설치된 PyTorch 버전 출력
print(torch.cuda.is_available())  # True가 나오면 GPU 사용 가능
```

```python
import torch
import numpy as np
from sklearn.neighbors import NearestNeighbors

# 랜덤 데이터 생성 (2000개의 은하)
num_nodes = 2000
coordinates = np.random.rand(num_nodes, 3)  # 위치 정보: [x, y, z]
velocity = np.random.rand(num_nodes, 3)     # 속도 정보: [vx, vy, vz]
mass = np.random.rand(num_nodes, 1)         # 질량 정보

# k-NN을 이용해 엣지 만들기
k = 16  # 각 은하마다 k개의 가까운 은하를 연결
knn = NearestNeighbors(n_neighbors=k)
knn.fit(coordinates)
edges = knn.kneighbors_graph(coordinates).toarray()

# PyTorch Geometric 데이터 형식으로 변환
from torch_geometric.data import Data

edge_index = torch.tensor(np.array(np.where(edges)), dtype=torch.long)
x = torch.tensor(np.hstack([coordinates, velocity, mass]), dtype=torch.float)
y = torch.tensor(mass, dtype=torch.float)  # 목표값: 질량

# 데이터를 .pt로 저장
data = Data(x=x, edge_index=edge_index, y=y)
torch.save(data, 'graphs.pt')

print("그래프 데이터가 생성되었습니다.")
```
